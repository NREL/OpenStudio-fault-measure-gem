```python
df_v1 = pd.read_csv(folder+"\\"+"in_v1_NoEconomizer.csv", index_col=0)
df_v2 = pd.read_csv(folder+"\\"+"in_v6_hardsized_noeconomizer_fm.csv", index_col=0)
df_v3 = pd.read_csv(folder+"\\"+"in_v6_hardsized_noeconomizer_pm.csv", index_col=0)
df_v4 = pd.read_csv(folder+"\\"+"in_v7_hardsized_economizer_fm.csv", index_col=0)
df_v5 = pd.read_csv(folder+"\\"+"in_v7_hardsized_economizer_pm.csv", index_col=0)
df_v6 = pd.read_csv(folder+"\\"+"in_v8_hardsized_economizer_override1.csv", index_col=0)
df_v7 = pd.read_csv(folder+"\\"+"in_v9_fixedmin.csv", index_col=0)
df_v8 = pd.read_csv(folder+"\\"+"in_v10_propmin.csv", index_col=0)
# df_v8 = pd.read_csv(folder+"\\"+"in_v8_hardsized_oa_override.csv", index_col=0)
```

```python
type_dict = ['temperature','airflow','ratio','power']
temperature = [
    'Environment:Site Outdoor Air Drybulb Temperature [C](TimeStep)',
    '9 ZONE PVAV SUPPLY INLET NODE:System Node Temperature [C](TimeStep)',
    '9 ZONE PVAV MIXED AIR NODE:System Node Temperature [C](TimeStep) '
]
airflow = [
    '9 ZONE PVAV:Air System Outdoor Air Mass Flow Rate [kg/s](TimeStep)',
    '9 ZONE PVAV:Air System Mixed Air Mass Flow Rate [kg/s](TimeStep)',
    '9 ZONE PVAV:Air System Outdoor Air Mechanical Ventilation Requested Mass Flow Rate [kg/s](TimeStep)',
]
ratio = [
    '9 ZONE PVAV:Air System Outdoor Air Economizer Status [](TimeStep)',
    '9 ZONE PVAV:Air System Outdoor Air Flow Fraction [](TimeStep)',
    '9 ZONE PVAV:Air System Outdoor Air Minimum Flow Fraction [](TimeStep)',
]
power = [
    '9 ZONE PVAV 2SPD DX CLG COIL 212KBTU/HR 9.5EER:Cooling Coil Electric Power [W](TimeStep)',
]
```

```python
fig = make_subplots(
            rows=4, 
            cols=1,
            shared_xaxes=True,
#             row_heights=[0.3, 0.7]
        )

row=1
for types in [temperature, airflow, ratio, power]:
    
    for variable in types:
        
        fig.add_trace(go.Scatter(
            y=df_v1[variable],
            mode='lines',
            name="NoEconomizer: "+variable,
        ),row=row, col=1)
        
        fig.add_trace(go.Scatter(
            y=df_v2[variable],
            mode='lines',
            name="NoEconomizer_HS_FM: "+variable,
        ),row=row, col=1)
        
        fig.add_trace(go.Scatter(
            y=df_v3[variable],
            mode='lines',
            name="NoEconomizer_HS_PM: "+variable,
        ),row=row, col=1)
                
        fig.add_trace(go.Scatter(
            y=df_v4[variable],
            mode='lines',
            name="NormalEconomizer_FM: "+variable,
        ),row=row, col=1)
        
        fig.add_trace(go.Scatter(
            y=df_v5[variable],
            mode='lines',
            name="NormalEconomizer_PM: "+variable,
        ),row=row, col=1)
        
        fig.add_trace(go.Scatter(
            y=df_v6[variable],
            mode='lines',
            name="OAOverride_1_MVRMF: "+variable,
        ),row=row, col=1)
        
        fig.add_trace(go.Scatter(
            y=df_v7[variable],
            mode='lines',
            name="Solution1_FM: "+variable,
            line=dict(width=4, dash='dot')
        ),row=row, col=1)
        
        fig.add_trace(go.Scatter(
            y=df_v8[variable],
            mode='lines',
            name="Solution2_PM: "+variable,
            line=dict(width=4, dash='dot')
        ),row=row, col=1)
        
    row+=1
        
fig.update_layout(
    width=1300,
    height=700,
    margin=dict(
        l=10,
        r=550,
        t=10,
        b=10,
    ),
    showlegend=True,
    plot_bgcolor='white',
    legend=dict(
        font=dict(
            size=8,
        ),
        yanchor="top",
        y=1.0,
        xanchor="left",
        x=1.0,
    ),
)

fig.update_yaxes(
    title_text="<b>Temperature [C]</b>",
    title_font_size = 10,
    row=1, col=1
)

fig.update_yaxes(
    title_text="<b>Air flow [kg/s]</b>",
    title_font_size = 10,
    row=2, col=1
)

fig.update_yaxes(
    title_text="<b>Signal/Ratio</b>",
    title_font_size = 10,
    row=3, col=1
)

fig.update_yaxes(
    title_text="<b>Power [W]</b>",
    title_font_size = 10,
    row=4, col=1
)

# fig.update_traces(opacity=0.8)


fig.show()
```
